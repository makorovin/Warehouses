```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
import os
import requests
import json
import time
import Levenshtein as L
import urllib as url
%matplotlib inline
```

```python
data = pd.read_excel('data_full_info_changed.xlsx')
```

```python
descr_list = list(data.full_description)
```

```python
descr_str_split_space = ''
sighns = [',','.','_','!','?',':',';','`','~','\\','/','+','=','*','#','@','^','%','$','№','(',')','"']
for descr in descr_list:
    descr_l = descr.lower()
    for n in range(10):
        descr_l = descr_l.replace(str(n),'')
    for s in sighns:
        descr_l = descr_l.replace(s,' ')
    for sp in [' '*i for i in range(2,10,1)]:
        descr_l = descr_l.replace(sp,' ')
    descr_str_split_space += ' '+descr_l
words = descr_str_split_space.split(' ')
words = np.array(words)[np.array([len(w)for w in words])>4]
```

```python
for w1 in words:
    for w2 in words:
        dist = L.distance(w1,w2)
        if 0<dist<3:
            print(w1,w2)
```
